# 5. Split data in training and testing

```python
#Before splitting, iam going to remove
data_num = data.drop(columns=['ocean_proximity'])
```

```python
# now i can split between train and test
# i import the library sklearn to add the split function
from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(data_num,test_size=0.2,random_state=42)
```

## separate features (x) and label (y)

```python
x_train = train_set.drop(columns=['median_house_value']) # remove the house value to keep the 8 columns
y_train = train_set['median_house_value'] # i just choose the house value
```

## data cleaning

```python
data.info() # total_bedrooms columns has empty values, we need to fix that

#option 1: Delete rows with no data
#Advantage: solve error. Problem: we lose data
x_train_1 = x_train.dropna(subset=['total_bedrooms'])

#option 2: Delete the entire column
#Advantage: we dont lose rows. problem: we lose an entire colum of data
x_train_2 = x_train.drop(columns=['total_bedrooms'])

#option 3: fill the empty values with something
#Advantage: I dont lose any data. Problem : Iam introducing artificial data
# How i avoid problems with artificial data? Introducing "MEDIAN" values

median = x_train['total_bedrooms'].median()
x_train['total_bedrooms'].fillna(median,inplace=True)

```

```python
x_train.info()
```

# Scaling and Standarizing the values

```python
x_train.head(3)
```

## MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x_train_prepared = scaler.fit_transform(x_train)
```

## Standarization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_prepared = scaler.fit_transform(x_train)
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train_prepared,y_train)
```

## coefficients of the regression model

```python
#linear regression coeff is an equation like (y=ax1+bx2....) or (y-ax+b)
lin_reg.coef_
```

# Measure performance of the Model: RMSE & MAE

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
#first we generate the predictions to measure the performance
predictions = lin_reg.predict(x_train_prepared)
# Now I apply my metrics to the real value and the predictions
lin_mae = mean_absolute_error(y_train,predictions)
lin_rmse = mean_squared_error(y_train,predictions)

print('RMSE:', lin_rmse)
print('MAE:', lin_mae)


```
